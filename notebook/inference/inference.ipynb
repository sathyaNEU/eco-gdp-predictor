```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
artifacts_path = os.getcwd().removesuffix('notebook\\inference')+'artifacts\\'
```

```python
result_df = pd.read_csv(artifacts_path + 'results.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=result_df,
    x='Year',
    y='gdp_ppp',
    hue='Entity',  # Different colors for each country
    palette='tab20',
    linewidth=2  # Line thickness
)

# Adding title and labels
plt.title('GDP (PPP) vs Year for Different Countries', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (PPP)', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.tight_layout()

# Show plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=result_df,
    x='Year',
    y='emp_pop_ratio',
    hue='Entity',  # Different colors for each country
    palette='tab20',
    linewidth=2  # Line thickness
)

# Adding title and labels
plt.title('Employment-Population Ratio vs Year for Different Countries', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Employment-Population Ratio', fontsize=12)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.tight_layout()

# Show plot
plt.show()
```
